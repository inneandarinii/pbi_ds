## 2.3 Checking Missing Values

```python
credit_card_balance.isna().sum()
```

```python
# Menghitung jumlah nilai hilang di setiap kolom
missing_values = credit_card_balance.isnull().sum()

# Menghitung persentase nilai hilang di setiap kolom
missing_percentage = (missing_values / len(credit_card_balance)) * 100

# Membuat DataFrame untuk menampilkan kolom dengan nilai hilang dan persentasenya
missing_data = pd.DataFrame({'Missing Values': missing_values, 'Percentage': missing_percentage})

# Menampilkan kolom yang memiliki nilai hilang (nilai hilang > 0)
missing_data = missing_data[missing_data['Missing Values'] > 0].sort_values(by='Percentage', ascending=False)

missing_data
```

```python
#Menyeleksi baris dalam DF 'credit_card_balance', dimana nilai kolom  'AMT_PAYMENT_CURRENT' adalah NaN (tidak ada).
credit_card_balance.loc[credit_card_balance.AMT_PAYMENT_CURRENT.isna()]
```

Mengisi nilai yang hilang dengan angka 0 dianggap sesuai karena kekosongan pada kolom saldo kartu kredit mungkin disebabkan oleh klien yang tidak memiliki kartu kredit atau kartu kredit yang tidak memiliki saldo.

```python
credit_card_balance_cleaned = credit_card_balance.fillna(0)
csv_file_path = 'C:/Users/Inne Andarini/Downloads/credit_card_balance_cleaned.csv'
credit_card_balance_cleaned.to_csv(csv_file_path, index=False)
```
