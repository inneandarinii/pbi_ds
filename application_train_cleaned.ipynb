## 2.3 Memeriksa dan Mengatasi Null Value

```python
pd.set_option('display.max_columns', None)
application_train.isna().sum()
```

```python
application_train.isnull().sum()
```

```python
# Menghitung jumlah nilai hilang di setiap kolom
missing_values = application_train.isnull().sum()

# Menghitung persentase nilai hilang di setiap kolom
missing_percentage = (missing_values / len(application_train)) * 100

# Membuat DataFrame untuk menampilkan kolom dengan nilai hilang dan persentasenya
missing_data = pd.DataFrame({'Missing Values': missing_values, 'Percentage': missing_percentage})

# Menampilkan kolom yang memiliki nilai hilang (nilai hilang > 0)
missing_data = missing_data[missing_data['Missing Values'] > 0].sort_values(by='Percentage', ascending=False)

missing_data
```

### Membuang kolom yang tidak diperlukan

```python
clean_flag = application_train.drop(columns=['FLAG_DOCUMENT_2', 'FLAG_DOCUMENT_3',
       'FLAG_DOCUMENT_4', 'FLAG_DOCUMENT_5', 'FLAG_DOCUMENT_6',
       'FLAG_DOCUMENT_7', 'FLAG_DOCUMENT_8', 'FLAG_DOCUMENT_9',
       'FLAG_DOCUMENT_10', 'FLAG_DOCUMENT_11', 'FLAG_DOCUMENT_12',
       'FLAG_DOCUMENT_13', 'FLAG_DOCUMENT_14', 'FLAG_DOCUMENT_15',
       'FLAG_DOCUMENT_16', 'FLAG_DOCUMENT_17', 'FLAG_DOCUMENT_18',
       'FLAG_DOCUMENT_19', 'FLAG_DOCUMENT_20', 'FLAG_DOCUMENT_21'])
```

```python
missing_percentage = clean_flag.isnull().mean()
```

### Membuang baris dengan nilai hilang lebih dari 10%

```python
columns_to_keep = missing_percentage[missing_percentage <= 0.1].index
clean_train = clean_flag[columns_to_keep]
```

### Mengisi nilai hilang pada kolom kategori dengan modus

```python
clean_train['NAME_TYPE_SUITE'].fillna(clean_train['NAME_TYPE_SUITE'].mode()[0], inplace=True)
```

### Mengisi nilai hilang pada kolom numerik dengan 0

```python
clean_train = clean_train.fillna(0)
```

### Memeriksa bentuk data sebelum dan sesudah data cleaning

```python
print(f'application train data before cleaning: {application_train.shape[0]} rows and {application_train.shape[1]} columns')
print(f'application train data after cleaning: {clean_train.shape[0]} rows and {clean_train.shape[1]} columns')
```
